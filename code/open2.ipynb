### 도수분포표

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, var in enumerate(category_list):
	sns.countplot(x = var, data = data, ax = axes[i])
	axes[i].set_title(f'Count Plot of {var}', fontsize=14)
	axes[i].set_xlabel(var, fontsize = 12)
	axes[i].set_ylabel('Count', fontsize = 12)
	if var == 'Category':
		axes[i].tick_params(axis='x', rotation = 90)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (12, 4))

for i, var in enumerate(numeric_list, start = 1):
	plt.subplot(1, 3, i)
	plt.hist(data[var], bins = 15, edgecolor = 'black')
	plt.title(f'{var}')
plt.tight_layout()
plt.show()
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df
```

```python
for var in numeric_list:
	remove_outliers_iqr(data, var)
```

```python
plt.figure(figsize = (12, 4))

for i, var in enumerate(numeric_list, start = 1):
	plt.subplot(1, 3, i)
	plt.hist(data[var], bins = 15, edgecolor = 'black')
	plt.title(f'{var}')
plt.tight_layout()
plt.show()
```
